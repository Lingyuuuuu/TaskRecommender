```python
# Function to generate user-task interaction
import numpy as np

def generate_user_task_interaction(users, tasks):
    interactions = []
    skip_percentage = 20
    for (user_index, user) in users.iterrows():
        for (task_index, task) in tasks.iterrows():

            rand_num = random.randint(0, 100)
            if rand_num < skip_percentage:
                pass

            else:
                if (user["Motivation"] == "Weight Management" and task["Type"] == "Physical") or \
                    (user["Motivation"] == "Improve Sleep" and task["Type"] == "Rest") or \
                    (user["Motivation"] == "Increase Energy" and task["Type"] == "Cardio") or \
                    (user["Motivation"] == "Reduce Stress" and task["Type"] == "Flexibility") or \
                    (user["Motivation"] == "Build Muscle" and task["Type"] == "Strength") or \
                    (user["Motivation"] == "Social Interaction" and task["Type"] == "Outdoor"):
                    
                    mean = 0.75
                    std_dev = 0.3
                    completion_prob = np.random.normal(mean, std_dev)

                else:
                    mean = 0.45
                    std_dev = 0.3
                
                    completion_prob = np.random.normal(mean, std_dev)


                if user["Overall Health Status"] == "Poor" and task["Intensity"] == "Low":
                    completion_prob += 0.2
                elif ((user["Overall Health Status"] == "Fair") or (user["Overall Health Status"] == "Good")) and (task["Intensity"] == "Medium"):
                    completion_prob += 0.2
                elif user["Overall Health Status"] == "Excellent" and task["Intensity"] == "High":
                    completion_prob += 0.2
                else:
                    completion_prob -= 0.2

                completion = completion_prob > 0.5  # True if clipped_num > 0.5, False otherwise               
                interactions.append({"taskid": task["taskid"], "userid": user["userid"], "completion": completion})
    return interactions


interactions = generate_user_task_interaction(user, task)

task_user_interaction = pd.DataFrame(interactions)
task_user_interaction = task_user_interaction.reset_index().rename(columns={'index': 'interaction_id'})
task_user_interaction['completion'] = task_user_interaction['completion'].astype(int)

task_user_interaction.to_csv("data_new/usertasks_new.csv", index=False)
```
